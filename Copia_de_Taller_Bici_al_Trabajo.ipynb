## **PREGUNTA 1**
¿La distribución de cnt puede tratarse como aproximadamente normal, realiza el histrograma y el gráfico Q-Q?

```python
import matplotlib.pyplot as plt
from scipy import stats

# Histogram
plt.figure(figsize=(14, 5))
plt.subplot(1, 2, 1) # 1 row, 2 columns, 1st plot
plt.hist(df['cnt'], bins=30, edgecolor='black')
plt.title('Histogram of Daily Bike Rentals (cnt)')
plt.xlabel('Daily Rentals')
plt.ylabel('Frequency')

# Q-Q plot
plt.subplot(1, 2, 2) # 1 row, 2 columns, 2nd plot
stats.probplot(df['cnt'], dist="norm", plot=plt)
plt.title('Q-Q Plot of Daily Bike Rentals (cnt)')

plt.tight_layout()
plt.show()
```

## **PREGUNTA 2**
¿Cómo co-varía cnt con temp, hum y windspeed? ¿La relación es lineal o solo monótona?. Ayuda: Calcular la correlación de Pearson y de Spearman.

```python
# Calcular los coeficientes de correlación de Pearson
# Esto mide la relación lineal entre las variables.
pearson_corr = df[['cnt', 'temp', 'hum', 'windspeed']].corr(method='pearson')
print("Correlación de Pearson:")
print(pearson_corr['cnt'][['temp', 'hum', 'windspeed']])

# Calcular los coeficientes de correlación de Spearman
# Esto mide la relación monótona (no necesariamente lineal) entre las variables.
spearman_corr = df[['cnt', 'temp', 'hum', 'windspeed']].corr(method='spearman')
print("\nCorrelación de Spearman:")
print(spearman_corr['cnt'][['temp', 'hum', 'windspeed']])
```

### **Interpretación pregunta 2**
Esto calcula dos tipos de correlación: Pearson para relaciones lineales y Spearman para relaciones monótonas. Esto nos permite ver si la conexión entre cnt (alquileres) y temp, hum, windspeed (clima) es directa y proporcional (lineal) o simplemente sigue una tendencia ascendente o descendente (monótona), sin ser necesariamente una línea recta.

## **PREGUNTA 3**
¿Desde qué tamaño muestral el TLC justifica intervalos y pruebas basados en normalidad para la media de cnt?

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 2 & 3: Simulate sampling distributions and calculate means
sample_sizes = [5, 10, 30, 50, 100, 200, 500]  # Different sample sizes to test
n_simulations = 1000  # Number of simulations for each sample size

sample_means = {size: [] for size in sample_sizes}

data = df['cnt'].values

for size in sample_sizes:
    for _ in range(n_simulations):
        sample = np.random.choice(data, size=size, replace=True)
        sample_means[size].append(np.mean(sample))

# Step 4: Visualize the distribution of sample means
plt.figure(figsize=(15, 10))
for i, size in enumerate(sample_sizes):
    plt.subplot(3, 3, i + 1)
    plt.hist(sample_means[size], bins=30, edgecolor='black')
    plt.title(f'Sample Size: {size}')
    plt.xlabel('Sample Mean of cnt')
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

### **Respuesta pregunta 3**
Al visualizar estos histogramas, podemos ver cómo la forma de la distribución de las medias muestrales se vuelve cada vez más parecida a una distribución normal a medida que el tamaño de la muestra aumenta. El punto donde la distribución se parece lo suficiente a una normal es el tamaño muestral a partir del cual el TLC justifica el uso de pruebas e intervalos basados en la normalidad para la media.

## **PREGUNTA 4**
En días laborales, ¿la media de cnt ≥ 4.500? (plantea y resuelve un test de hipótesis de una muestra).

```python
# 1. Filtrar datos de días laborales
cnt_laborales = df[df['workingday'] == 1]['cnt']

# Valor de referencia para la prueba
valor_hipotesis = 4500

# Test de hipótesis: ¿media de cnt_laborales >= 4500?
from scipy.stats import ttest_1samp

# Ejecutamos t-test
t_stat, p_val_bilateral = ttest_1samp(cnt_laborales, valor_hipotesis)

# Como la prueba es unilateral (H1: media > 4500):
if t_stat > 0:
    p_val_unilateral = p_val_bilateral / 2
else:
    p_val_unilateral = 1 - (p_val_bilateral / 2)

print(f"Media muestral = {cnt_laborales.mean():.2f}")
print(f"T-Statistic = {t_stat:.4f}")
print(f"P-valor (unilateral) = {p_val_unilateral:.4f}")

# Decisión con α=0.05
alpha = 0.05
if p_val_unilateral < alpha:
    print("✅ Rechazamos H0: la media de cnt en días laborales es significativamente mayor a 4500.")
else:
    print("❌ No podemos rechazar H0: no hay evidencia suficiente de que la media sea mayor a 4500.")
```

### **Interpretación pregunta 4**
El análisis estadístico formal respalda si la meta de 4.500 viajes diarios es alcanzada o no. No basta con observar promedios: se necesita verificar que la diferencia sea estadísticamente confiable.

## **PREGUNTA 5**
¿Cuál es un IC bootstrap para la mediana de cnt y cómo cambia tu mensaje si la distribución es sesgada?

```python
# Número de remuestreos
n_iterations = 5000
sample_size = len(df['cnt'])
medians = []

import numpy as np
import matplotlib.pyplot as plt

# Parámetros bootstrap
n_iterations = 5000
sample = df['cnt'].values
n = len(sample)
rng = np.random.default_rng(123)  # reproducibilidad

# Remuestreo bootstrap: calcular medianas
medians = np.empty(n_iterations)
for i in range(n_iterations):
    idx = rng.integers(0, n, size=n)   # muestreo con reemplazo
    medians[i] = np.median(sample[idx])

# IC percentil 95% (2.5% - 97.5%)
alpha = 0.05
lower, upper = np.percentile(medians, [100*alpha/2, 100*(1-alpha/2)])
# Estadísticos
median_observada = np.median(sample)
mean_boot = medians.mean()
bias = mean_boot - median_observada

print(f"Mediana observada (muestra): {median_observada:.2f}")
print(f"Mediana bootstrap (promedio remuestreos): {mean_boot:.2f}")
print(f"Bias estimado (bootstrap - observada): {bias:.2f}")
print(f"IC bootstrap percentil 95% para la mediana: [{lower:.2f}, {upper:.2f}]")

# Gráfico de la distribución bootstrap de las medianas
plt.figure(figsize=(8,4))
plt.hist(medians, bins=40, density=True, alpha=0.7)
plt.axvline(median_observada, color='black', linestyle='--', label='Mediana observada')
plt.axvline(lower, color='red', linestyle=':', label=f'IC 2.5% = {lower:.0f}')
plt.axvline(upper, color='red', linestyle=':', label=f'IC 97.5% = {upper:.0f}')
plt.title('Distribución bootstrap de la mediana de cnt')
plt.xlabel('Mediana de cnt')
plt.ylabel('Densidad')
plt.legend()
plt.show()
```

**Interpretación pregunta 5**
La distribución de cnt es sesgada, por lo que la mediana es un mejor resumen que la media, ya que no se ve tan afectada por valores extremos.

Cuando la distribución no es simétrica, la mediana comunica mejor el “uso típico” de la bicicleta. Para informes públicos, reportar la mediana evita dar una visión distorsionada de la realidad.
